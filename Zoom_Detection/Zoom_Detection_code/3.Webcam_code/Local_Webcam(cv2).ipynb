### 동영상 실시간 캡쳐 Try
- 실시간 웹캠으로 인물 탐지가 불가능할 때,
- 차선책으로 동영상 녹화본 인물탐지를 위한 용도
- 동영상 => 이미지로 : 실시간 단위로 캡쳐가 된다면, 그 캡쳐 이미지를 인물탐지

```python
# step1.opencv 라이브러리 불러오기
import cv2

# step2.동영상 파일 열기
cap = cv2.VideoCapture('KakaoTalk_20221222_155015780.mp4')

# step3.무한 반복
while True:
    # 동영상 파일 존재 여부(True/False)와 현재 프레임 이미지를 읽음
    retval, frame = cap.read()
    
    # 만약 동영상 파일이 존재하지 않으면 while 반복문 종료
    if retval == False:
        break
    
    key = cv2.waitKey(30) # 30ms 동안 키입력 대기
    if key == ord('c'):
        cv2.imwrite("capture1.png", frame)

    # 10초가 지나거나, ESC 키가 입력되면 while 반복문 종료
    if cv2.waitKey(10) == 27:
        break
        
    # 'frame'이란 창 이름으로 현재 프레임 출력
    cv2.imshow('frame', frame)

# step4.동영상 파일 닫고 모든창 종료
cap.release()

cv2.destroyAllWindows()
```

- 위 코드 실행 시, c 키를 입력했을 때 아래 이미지와 같이 캡쳐 본 저장
